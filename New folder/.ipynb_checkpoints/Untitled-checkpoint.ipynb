```python
from geopy.extra.rate_limiter import RateLimiter
from geopy.geocoders import Nominatim

locator = Nominatim(user_agent="myGeocoder")

# 1 - conveneint function to delay between geocoding calls
geocode = RateLimiter(locator.geocode, min_delay_seconds=1)
# 2- - create location column
inpt_new['geocode'] = inpt_new['Location'].apply(geocode)
# 3 - create longitude, laatitude and altitude from location column (returns tuple)
inpt_new['point'] = inpt_new['geocode'].apply(lambda loc: tuple(loc.point) if loc else None)
# 4 - split point column into latitude, longitude and altitude columns
inpt_new[['Latitude', 'Longitude', 'Altitude']] = pd.DataFrame(inpt_new['point'].tolist(), index=inpt_new.index)
```

```python
import json
 
def write_json(new_data, filename='df2.json'):
    with open(filename,'r+') as file:
          # First we load existing data into a dict.
        file_data = json.load(file)
        # Join new_data with file_data inside emp_details
        file_data["data"].append(new_data)
        # Sets file's current position at offset.
        file.seek(0)
        # convert back to json.
        json.dump(file_data, file, indent = 4)
 
```
